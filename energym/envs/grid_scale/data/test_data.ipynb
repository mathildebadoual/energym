```python
import pytz
file_gen = open('gen_caiso.pkl', 'rb')

data_gen = pickle.load(file_gen)
data_gen = data_gen.sort_values('timestamp')
data_gen = data_gen.set_index('timestamp')

data_other = data_gen[data_gen['fuel_name'] == 'other']
data_wind = data_gen[data_gen['fuel_name'] == 'wind']
data_solar = data_gen[data_gen['fuel_name'] == 'solar']

numhours = 29518
for date in [datetime.datetime(2015, 6, 6, 8, 0, 0, tzinfo=pytz.utc) + datetime.timedelta(hours=x) for x in range(0, numhours)]:
    try:
        data_other.loc[date]
    except:
        data_other = data_other.append(pd.DataFrame.from_dict({date: []}, orient='index'),
            sort=True)
    try:
        data_wind.loc[date]
    except:
        data_wind = data_wind.append(pd.DataFrame.from_dict({date: []}, orient='index'),
            sort=True)
    try:
        data_solar.loc[date]
    except:
        data_solar = data_solar.append(pd.DataFrame.from_dict({date: []}, orient='index'),
            sort=True)

data_gen = data_gen.sort_index()
data_gen = data_gen.interpolate()

data_gen[data_gen['fuel_name'] == 'other'] = data_other
data_gen[data_gen['fuel_name'] == 'wind'] = data_wind
data_gen[data_gen['fuel_name'] == 'solar'] = data_solar

data_gen = data_gen.loc[:datetime.datetime(2018, 10, 13, 6, 0, 0, tzinfo=pytz.utc)]
data_gen = data_gen.reset_index()
data_gen = data_gen.rename(columns={'index': 'timestamp'})

print(data_gen)
pickle.dump(data_gen, open("gen_caiso.pkl", "wb" ))
```
